```python
def create_binary_matrix(df):
    df_bin = df.copy()
    cols_to_bin = df_bin.columns
    df_bin[cols_to_bin] = (df_bin[cols_to_bin] != 0).astype(int)
    return df_bin
```

```python
def filter_rare_side_effects(df, min_count=15):
    cols = df.columns[1:]
    counts = df[cols].sum(axis=0)
    filtered_cols = [df.columns[0]] + counts[counts >= min_count].index.tolist()
    return df[filtered_cols]
```

```python

lltpt_binary = create_binary_matrix(drug_sideeffect_lltpt_clean)
hlt_binary = create_binary_matrix(drug_sideeffect_hlt_clean)
hlgt_binary = create_binary_matrix(drug_sideeffect_hlgt_clean)

lltpt_binary = filter_rare_side_effects(lltpt_binary, min_count=15)
hlt_binary = filter_rare_side_effects(hlt_binary, min_count=15)
hlgt_binary = filter_rare_side_effects(hlgt_binary, min_count=15)


lltpt_binary.to_csv('lltpt_binary.csv', index=False)
hlt_binary.to_csv('hlt_binary.csv', index=False)
hlgt_binary.to_csv('hlgt_binary.csv', index=False)

drug_sideeffect_lltpt_clean.to_csv('lltpt_quant.csv', index=False)
drug_sideeffect_hlt_clean.to_csv('hlt_quant.csv', index=False)
drug_sideeffect_hlgt_clean.to_csv('hlgt_quant.csv', index=False)

pd.Series(smiles_list_new).to_csv('smiles.csv', index=False, header=False)

```
